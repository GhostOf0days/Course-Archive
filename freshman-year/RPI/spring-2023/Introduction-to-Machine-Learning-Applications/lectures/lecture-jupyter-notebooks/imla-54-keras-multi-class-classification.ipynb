```python
# Read and preprocess the data
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

raw_df = pd.read_csv('https://raw.githubusercontent.com/lmanikon/lmanikon.github.io/master/teaching/datasets/bostonHousing.csv')
raw_df.head()

# Drop null rows
non_null_df = raw_df.dropna()
non_null_df['MEDV'].describe()

X = non_null_df.drop(columns="MEDV")
# Discretize the target variable so this becomes a classification problem
y = np.digitize(non_null_df['MEDV'], np.array([5.0, 21.2, 68.0]))

X_scaled = pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns)
X_scaled

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state = 25, test_size=0.2)
```

```python
# Build the network

from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
# Add layers
model.add(Dense(12, input_dim=13, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# compile the keras model with binary crossentropy and optimizer with a certain learning rate
#Adam optimization is a stochastic gradient descent method that is based on adaptive estimation of first-order and second-order moments
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit the keras model on the dataset
model.fit(X_train, y_train, epochs=150, batch_size=10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X_test, y_test)
print('Keras accuracy: %.2f' % accuracy)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
print("Scikit-learn MLPClassifier accuracy: ", accuracy_score(y_test, MLPClassifier(random_state=1, max_iter=5).fit(X_train, y_train).predict(X_test)))
```
