```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

raw_df = pd.read_csv('https://raw.githubusercontent.com/lmanikon/lmanikon.github.io/master/teaching/datasets/KaggleStroke.csv')
raw_df.head()
non_null_df = raw_df.dropna()
X = StandardScaler().fit_transform(pd.get_dummies(non_null_df.drop(columns=['id', 'stroke'])))
y = non_null_df['stroke']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 25, test_size=0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Build the model
model = Sequential()
model.add(Dense(12, input_dim=21, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=150, batch_size=10)
```

```python
_, accuracy = model.evaluate(X_test, y_test)
print('Keras accuracy: %.2f' % accuracy)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
print("Scikit-learn MLPClassifier accuracy: ", accuracy_score(y_test, MLPClassifier(random_state=1, max_iter=5).fit(X_train, y_train).predict(X_test)))
```
