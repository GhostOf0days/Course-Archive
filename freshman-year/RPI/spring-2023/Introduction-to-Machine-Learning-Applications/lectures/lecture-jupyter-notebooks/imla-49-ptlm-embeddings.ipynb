# Using pre-trained language models for text vectorization

This notebook will use the [SentenceTransformers](https://www.sbert.net/) library to compute sentences to embeddings, which we can then use in standard Scikit-Learn models.

It is adapted from the example https://github.com/UKPLab/sentence-transformers/blob/master/examples/applications/clustering/kmeans.py

K-Means is memory-intensive. You may need to run this in Google Colab.

```python
# Corpus with example sentences
corpus = ['A man is eating food.',
          'A man is eating a piece of bread.',
          'A man is eating pasta.',
          'The girl is carrying a baby.',
          'The baby is carried by the woman',
          'A man is riding a horse.',
          'A man is riding a white horse on an enclosed ground.',
          'A monkey is playing drums.',
          'Someone in a gorilla costume is playing a set of drums.',
          'A cheetah is running behind its prey.',
          'A cheetah chases prey on across a field.'
          ]
```

```python
from sentence_transformers import SentenceTransformer
embedder = SentenceTransformer('all-MiniLM-L6-v2')  # Could use another pre-trained language model, such as 'distilbert-base-nli-mean-tokens'
X = embedder.encode(corpus)
```

```python
X
```

```python
assert len(X) == len(corpus)
len(X[0])
```

```python
len(X[1])
```

```python
# Cluster the sentences
from sklearn.cluster import KMeans
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)

clustered_sentences = [[] for i in range(num_clusters)]
for sentence_id, cluster_id in enumerate(kmeans.labels_):
    clustered_sentences[cluster_id].append(corpus[sentence_id])

for i, cluster in enumerate(clustered_sentences):
    print("Cluster ", i + 1)
    print(cluster)
    print("")
```
